```python
contents = urllib.request.urlopen(
    "https://api.openligadb.de/getavailableleagues"
).read()
data = json.loads(contents)
data
```

```python
league_season_list = []
for game in data:
    league_season_list.append((game["leagueShortcut"], int(game["leagueSeason"])))
```
